```python
from evaluate import load

# Sample predictions and references
predictions = ["The cat sat on the mat."]
references = ["The cat is sitting on the mat."]

# ROUGE
rouge = load("rouge")
rouge_result = rouge.compute(predictions=predictions, references=references)
print("ROUGE:", rouge_result)

# BLEU
bleu = load("bleu")
bleu_result = bleu.compute(predictions=predictions, references=references)
print("BLEU:", bleu_result)

# METEOR
meteor = load("meteor")
meteor_result = meteor.compute(predictions=predictions, references=references)
print("METEOR:", meteor_result)
```

```python
from bert_score import score

predictions = ["The cat sat on the mat."]
references = ["The cat is sitting on the mat."]

P, R, F1 = score(predictions, references, lang="en", rescale_with_baseline=True)
print(f"BERTScore - Precision: {P.mean():.4f}, Recall: {R.mean():.4f}, F1: {F1.mean():.4f}")
```

```python
from moverscore import get_idf_dict, word_mover_score

references = ["The cat is sitting on the mat."]
hypotheses = ["The cat sat on the mat."]

idf_hyp = get_idf_dict(hypotheses)
idf_ref = get_idf_dict(references)

scores = word_mover_score(references, hypotheses, idf_ref, idf_hyp, stop_words=[], n_gram=1, remove_subwords=True)
print(f"MoverScore: {scores}")
```

```python
from bleurt import score

references = ["The cat is sitting on the mat."]
candidates = ["The cat sat on the mat."]

scorer = score.BleurtScorer()
scores = scorer.score(references=references, candidates=candidates)
print(f"BLEURT Score: {scores}")
```
